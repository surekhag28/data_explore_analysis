```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.head()
```

```python
df = df[['active','cured','death','new_active','new_cured','new_death','new_positive','positive','state_code','state_name']]
df = df.iloc[:-1,:]
df
```

### Total positive cases based on states of India

```python
fig = plt.figure(figsize=(10,10))
plt.barh(df['state_name'],df['positive'],color='g')
plt.title('Covid positive cases based in states')
plt.xlabel('Covid postive count')
plt.ylabel('States')
plt.show()
```

### Active number of cases based on state

```python
fig = plt.figure(figsize=(10,10))
plt.barh(df['state_name'],df['active'],color='y')
plt.title('Covid active cases based in states')
plt.xlabel('Covid active count')
plt.ylabel('States')
plt.show()
```

### Death number of cases based on state

```python
fig = plt.figure(figsize=(10,10))
plt.barh(df['state_name'],df['death'],color='r')
plt.title('Covid death cases based in states')
plt.xlabel('Covid death count')
plt.ylabel('States')
plt.show()
```
